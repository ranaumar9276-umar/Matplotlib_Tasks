```python
#TASK 1 

import matplotlib.pyplot as plt

semester = [1, 2, 3, 4, 5, 6]
python_marks = [65, 72, 78, 84, 90, 96]
numpy_marks = [0, 0, 0, 0, 82, 94]
pandas_marks = [0, 0, 0, 0, 80, 92]

plt.figure(figsize=(10,6))
plt.plot(
    semester,
    python_marks,
    marker="o",
    linewidth=3,
    color="blue",
    label="Python"
)
plt.plot(
    semester,
    numpy_marks,
    marker="s",
    linewidth=3,
    color="green",
    label="NumPy"
)
plt.plot(
    semester,
    pandas_marks,
    marker="^",
    linewidth=3,
    color="red",
    label="Pandas"
)
plt.title("Programming Skills Growth",fontsize=16)
plt.xlabel("Semester")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()

plt.show()
```

```python
# TASK  2

import matplotlib.pyplot as plt

years = [2021,2022,2023,2024,2025]
ai = [8,14,22,35,48]
machine_learning = [6,12,18,28,40]
data_science = [5,10,17,25,37]

plt.figure(figsize=(10,6))
plt.plot(years,ai,
         marker="o",
         linewidth=3,
         label="Artificial Intelligence")

plt.plot(years,machine_learning,
         marker="s",
         linewidth=3,
         label="Machine Learning")

plt.plot(years,data_science,
         marker="^",
         linewidth=3,
         label="Data Science")

plt.title("Research Publications Comparison")
plt.xlabel("Years")
plt.ylabel("Research Papers")
plt.grid(True)
plt.legend()

plt.show()
```
