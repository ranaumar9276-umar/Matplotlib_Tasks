```python
#Task 1 

import matplotlib.pyplot as plt
semester = [1, 2, 3, 4, 5, 6]
python_marks = [65, 72, 78, 84, 90, 96]

plt.plot(semester, python_marks)
plt.title("Python Learning Progress")
plt.xlabel("Semester")
plt.ylabel("Python Marks")

plt.show()
```
