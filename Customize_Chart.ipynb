```python
# TASK 1
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
attendance = [55, 63, 81, 74, 96, 108]

plt.figure(figsize=(10,5))
plt.plot(
    months,
    attendance,
    color="blue",
    linestyle="--",
    marker="o",
    linewidth=3,
    markersize=8,
    label="Students"
)
plt.title("AI Workshop Attendance", fontsize=16)
plt.xlabel("Months", fontsize=12)
plt.ylabel("Students", fontsize=12)
plt.grid(True)
plt.legend()

plt.show()
```

```python
#TASK 2

import matplotlib.pyplot as plt

weeks = ["Week 1","Week 2","Week 3","Week 4","Week 5","Week 6"]
completion = [10,28,45,63,82,100]
plt.figure(figsize=(10,5))
plt.plot(
    weeks,
    completion,
    color="green",
    marker="s",
    linestyle="-.",
    linewidth=3,
    markersize=9,
    label="Course Completion"
)
plt.title("Machine Learning Course Progress")
plt.xlabel("Weeks")
plt.ylabel("Completion (%)")
plt.ylim(0,110)
plt.grid(True)
plt.legend()

plt.show()
```
