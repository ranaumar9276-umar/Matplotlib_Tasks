```python
# TASK 1

import matplotlib.pyplot as plt

semester = [1, 2, 3, 4, 5, 6]
python_marks = [65, 72, 78, 84, 90, 96]
plt.figure(figsize=(10,6))
plt.plot(
    semester,
    python_marks,
    marker="o",
    color="blue",
    linewidth=3,
    label="Python Marks"
)
plt.title("Student Python Learning Progress", fontsize=16)
plt.xlabel("Semester")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()

plt.savefig("student_python_progress.png", dpi=300)

plt.show()
```

```python
# TASK 2

import matplotlib.pyplot as plt

years = [2021, 2022, 2023, 2024, 2025]
papers = [15, 24, 36, 49, 63]

plt.figure(figsize=(10,6))
plt.plot(
    years,
    papers,
    marker="s",
    color="green",
    linewidth=3,
    label="Research Papers"
)
plt.title("AI Research Publications")
plt.xlabel("Year")
plt.ylabel("Number of Papers")
plt.grid(True)
plt.legend()

# Save Chart
plt.savefig("ai_research_publications.jpg", dpi=300)

plt.show()
```
