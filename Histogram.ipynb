```python
# TASK 1

import matplotlib.pyplot as plt

python_marks = [
    55,60,62,65,67,69,70,72,74,75,
    76,78,79,80,82,84,85,86,88,90,
    91,92,93,94,95,96,97,98,99,100
]
plt.figure(figsize=(10,6))
plt.hist(
    python_marks,
    bins=10,
    color="skyblue",
    edgecolor="black"
)
plt.title("Distribution of Python Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.grid(axis="y")

plt.show()
```
