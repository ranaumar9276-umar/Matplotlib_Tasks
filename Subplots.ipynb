```python
# TASK 1

import matplotlib.pyplot as plt

semester = [1,2,3,4,5,6]
python_marks = [65,72,78,84,90,96]
cgpa = [2.90,3.10,3.25,3.40,3.55,3.72]
attendance = [82,85,88,90,92,95]
internship = [5,8,12,18,25,35]

plt.figure(figsize=(12,8))

# Graph 1
plt.subplot(2,2,1)
plt.plot(semester,python_marks,marker="o")
plt.title("Python Progress")

# Graph 2
plt.subplot(2,2,2)
plt.bar(semester,cgpa)
plt.title("Semester CGPA")

# Graph 3
plt.subplot(2,2,3)
plt.hist(attendance,bins=5)
plt.title("Attendance Distribution")

# Graph 4
plt.subplot(2,2,4)
plt.plot(semester,internship,marker="s")
plt.title("Internship Growth")
plt.tight_layout()

plt.show()
```

```python
# TASK 2

import matplotlib.pyplot as plt

years = [2021,2022,2023,2024,2025]
research = [15,24,35,48,62]
citations = [100,180,260,370,520]
funding = [2,3,5,7,10]
downloads = [1200,1800,2600,3800,5200]

plt.figure(figsize=(12,8))

# Line Chart
plt.subplot(2,2,1)
plt.plot(years,research,marker="o")
plt.title("Research Papers")

# Bar Chart
plt.subplot(2,2,2)
plt.bar(years,citations)
plt.title("Citations")

# Scatter Plot
plt.subplot(2,2,3)
plt.scatter(funding,citations,s=100)
plt.title("Funding vs Citations")

# Pie Chart
plt.subplot(2,2,4)
plt.pie(
downloads,
labels=years,
autopct="%1.1f%%",
startangle=90
)
plt.title("Downloads Distribution")
plt.tight_layout()

plt.show()
```
