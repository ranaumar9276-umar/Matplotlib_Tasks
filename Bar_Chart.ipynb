```python
# TASK 1

import matplotlib.pyplot as plt

department = [
    "Software Engineering",
    "Computer Science",
    "Artificial Intelligence",
    "Cyber Security",
    "Data Science"
]
cgpa = [3.45,3.61,3.78,3.52,3.82]
plt.figure(figsize=(10,6))
plt.bar(
    department,
    cgpa
)
plt.title("Department Wise Average CGPA")
plt.xlabel("Departments")
plt.ylabel("CGPA")
plt.xticks(rotation=20)
plt.grid(axis="y")

plt.show()
```

```python
# TASK 2

import matplotlib.pyplot as plt

languages = [
    "Python",
    "JavaScript",
    "Java",
    "C++",
    "Go",
    "Rust"
]
popularity = [98,90,82,78,65,58]
plt.figure(figsize=(10,6))
plt.bar(
    languages,
    popularity
)
plt.title("Most Popular Programming Languages")
plt.xlabel("Languages")
plt.ylabel("Popularity Score")
plt.grid(axis="y")

plt.show()
```
