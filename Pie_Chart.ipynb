```python
# TASK 1

import matplotlib.pyplot as plt

status = [
    "Selected",
    "Rejected",
    "Waiting"
]
students = [
    45,
    35,
    20
]
colors = [
    "green",
    "red",
    "orange"
]
plt.figure(figsize=(7,7))
plt.pie(
    students,
    labels=status,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("AI Internship Application Status")

plt.show()
```

```python
# TASK 2

import matplotlib.pyplot as plt

research_area = [
    "Artificial Intelligence",
    "Machine Learning",
    "Deep Learning",
    "Data Science",
    "Computer Vision"
]
papers = [
    30,
    25,
    18,
    15,
    12
]
colors = [
    "gold",
    "skyblue",
    "lightgreen",
    "orange",
    "violet"
]
plt.figure(figsize=(8,8))
plt.pie(
    papers,
    labels=research_area,
    colors=colors,
    autopct="%1.1f%%",
    shadow=True,
    startangle=140
)
plt.title("Research Papers by Area")

plt.show()
```
